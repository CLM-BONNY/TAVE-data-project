```python
import pandas as pd
# 제주 내국인 관관객 현황 데이터(2015 ~ 2020)

population = pd.read_excel("/content/drive/MyDrive/제주특별자치도_내국인관광객현황(2015.01~2020.12).xlsx")
population
```

```python
# 제주 내국인 관광객 현황 데이터: 형태별(개별여행), 형태별(부분 패키지), 형태별(패키지) 컬럼 값 병합

population["내국인 관광객 수"] = population[["행태별 - 개별여행", "행태별 - 부분패키지", "행태별 - 패키지"]].sum(axis=1)
population = population[["구분연월", "내국인 관광객 수"]]

population
```

```python
# 제주도 날씨 데이터(2015 ~ 2020)
weather = pd.read_excel("/content/drive/MyDrive/제주도지역_기상개황정보(2015.01~2020.12).xlsx")
weather
```

```python
# 제주도 날씨 데이터와 제주 내국인 관광객수 총합 데이터 병합

data = pd.concat([weather, population["내국인 관광객 수"]], axis=1)
data
```

```python
# '최심신적설'컬럼 -> 결측값이 너무 많이 존재하지 하기 때문에 drop

data.drop(columns=["최심신적설 (cm)"], inplace=True)
data
```

```python
# 컬럼 내용 확인하기

data.info()
```

- "평균운량 (1/10)" 컬럼에 -, 0으로 채워진 행이 많아 결측값 채우는 함수 필요
- 2015년, 2019년, 2020년 데이터는 있으므로 3년 각 월의 평균값으로 채워줄 예정

```python
data["월"] = 0
for i in range(len(data["내용"])):
    data["월"][i] = data["내용"][i] % 100
data
```

```python
# 최종 결측값 채우기 함수
def fill(column):
  # 평균 구하기
  meanList = [0 for i in range(12)]

  for i in range(len(data[column])):
    if data[column][i] == "-":
      data[column][i] = 0
    for j in range(12):
      if data["월"][i] == j+1:
        meanList[j] = meanList[j] + data[column][i]
  meanList = [round(x/3, 1) for x in meanList]

  # 평균으로 채우기
  for i in range(len(data[column])):
    if data[column][i] == 0:
      for j in range(12):
        if data["월"][i] == j+1:
          data[column][i] = meanList[j]

```

```python
fill("평균운량 (1/10)")
data["평균운량 (1/10)"][13:60]
```

```python
# Ridge Regression(평균 기온, 이슬점 온도)을 위한 라이브러리
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_percentage_error
```

```python
# Ridge Regression(평균기온 (℃), 이슬점온도 (℃))

features = ["평균기온 (℃)", "이슬점온도 (℃)"]
X = data[features]
y = data["내국인 관광객 수"]

# train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=97)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# 릿지 회귀분석 시행 (alpha의 기본값; 1)
ridge = Ridge().fit(X_train, y_train)
y_pred = ridge.predict(X_test)

print(f"RMSE: {mean_squared_error(y_test, y_pred, squared=False)}")
print(f"R2 Score: {r2_score(y_test, y_pred)}")
print(f"MAPE: {mean_absolute_percentage_error(y_test, y_pred)}")
```

- RMSE: 값이 낮을수록 좋다.
- R2 Score: 1에 가까울수록 좋다.
- MAPE: 값이 낮을수록 좋다. 다른 모델과 에러율 비교가 쉬움
